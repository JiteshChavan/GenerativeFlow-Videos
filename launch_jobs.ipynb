```python
JOB_NAME = "generative-flow"
INSTANCE_TYPE = "ml.p4d.24xlarge"
```

```python
import boto3
from sagemaker.core.helper.session_helper import Session, get_execution_role
from sagemaker.train.model_trainer import ModelTrainer, Mode
from sagemaker.train.configs import SourceCode, Compute
from sagemaker.core.shapes import StoppingCondition
from sagemaker.core import image_uris

role = get_execution_role()
sess = Session()
region = sess.boto_region_name


training_image = image_uris.retrieve(
framework="pytorch",
region=region,
version="2.6.0",
py_version="py312",
instance_type=INSTANCE_TYPE,
image_scope="training"
)

role = get_execution_role()
sess = Session()

source_code = SourceCode(
source_dir='.',
requirements="req.txt",
command=f"""bash ./init.sh""",
)

compute = Compute(
instance_type=INSTANCE_TYPE,
instance_count=1,
volume_size_in_gb=100
)

stopping_condition = StoppingCondition(max_runtime_in_seconds=2419200)

# Instead of Estimator, you could use ModelTrainer
trainer = ModelTrainer(
base_job_name=JOB_NAME,
sagemaker_session=sess,
training_image=training_image,
compute=compute,
stopping_condition=stopping_condition,
role=role,
source_code=source_code,
training_mode = Mode.SAGEMAKER_TRAINING_JOB
)
```

```python
trainer.train(wait=False)
```
